```python
#Get news from CNN
from newspaper import Article

url = 'https://www.cnn.com/2022/04/11/success/negotiate-job-offer-feseries/index.html'

article = Article(url)
article.download()
article.parse()

print("Authors: " + str(article.authors))
print("publish date: " + str(article.publish_date))
print("body:")
print(article.text)

text = article.text
```

```python
#Download en_core_web_md for spacy if not exists
#!python -m spacy download en_core_web_md
```

```python
#Load libs
import spacy

nlp = spacy.load('en_core_web_md')
```

```python
doc = nlp(text)

for sentence in doc.sents:
    print(str(sentence))
```

```python
import re

#Calculate r_i and A_i
pi = dict()
sentence_count = 0
for sentence in doc.sents:
    first_word = str(sentence).split(" ")[0]
    first_word = re.sub('\\n\\n', '', first_word)
    count = pi.setdefault(first_word, 0)
    pi[first_word] = count + 1
    sentence_count = sentence_count + 1

print(pi)
print("After \n")
for key, value in pi.items():
    pi[key] = value / sentence_count

print(pi)

print("######\n")
#Calculate state transition makers
a = dict()
for sentence in doc.sents:
    words = str(sentence).split(" ")
    for i in range(1, len(words)):
        key = '{}->{}'.format(words[i - 1], words[i])
        count = a.setdefault(key, 0)
        a[key] = count + 1
    
for key,value in a.items():
    a[key] = value / sentence_count

print(a)
```
